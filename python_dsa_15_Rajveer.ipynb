```python
#Q2 coin change

from typing import List

def coin_change(coins: List[int], amount: int) -> int:
    INF = float('inf')
    dp = [INF] * (amount + 1)
    dp[0] = 0
    
    for i in range(1, amount + 1):
        for coin in coins:
            if coin <= i:
                dp[i] = min(dp[i], 1 + dp[i - coin])
    
    return dp[amount] if dp[amount] != INF else -1
print(f"Input: coins=[1, 2, 5], amount=11, Output: {coin_change([1, 2, 5], 11)}")
print(f"Input: coins=[2], amount=3, Output: {coin_change([2], 3)}")
    
```

```python
#Q3.zero sum triplet 
def find_zero_sum_triplets(nums: list[int]) -> list[list[int]]:
    result = []
    nums.sort()  # pehle sort karo
    
    for i in range(len(nums) - 2):  # last 2 positions left aur right ke liye chhod do
        # duplicate i skip karo
        if i > 0 and nums[i] == nums[i-1]:
            continue
        
        left = i + 1
        right = len(nums) - 1
        
        while left < right:
            total = nums[i] + nums[left] + nums[right]
            
            if total == 0:
                result.append([nums[i], nums[left], nums[right]])
                
                # duplicates skip karo left side se
                while left < right and nums[left] == nums[left + 1]:
                    left += 1
                # duplicates skip karo right side se
                while left < right and nums[right] == nums[right - 1]:
                    right -= 1
                
                left += 1
                right -= 1
            
            elif total < 0:
                left += 1
            else:  # total > 0
                right -= 1
    
    return result
print(f"Input: [-1, 0, 1, 2, -1, -4], Output: {find_zero_sum_triplets([-1, 0, 1, 2, -1, -4])}")
print(f"Input: [0, 1, 1], Output: {find_zero_sum_triplets([0, 1, 1])}")
```

```python
#Q4 Sum of length if its Prime Number
def sum_if_length_prime(arr: list[int]) -> int:
     # START OF YOUR CODE
    # Task: Implement the logic to sum elements if the array's length is prime, otherwise return 0.
    # Store the final answer in the 'result' variable.
    n = len(arr)
    if n < 2:
        result = 0
    else:
        is_prime = True
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                is_prime = False
                break
        result = sum(arr) if is_prime else 0
    return result


# Example calls for testing
print(f"Input: [1, 2, 3], Output: {sum_if_length_prime([1, 2, 3])}")
print(f"Input: [10, 20, 30, 40], Output: {sum_if_length_prime([10, 20, 30, 40])}")
```

```python
#Q5 Binary TREE
class TreeNode:
  def __init__(self, val=0, left=None, right=None):
    self.val = val
    self.left = left
    self.right = right

def is_valid_bst(root: TreeNode) -> bool:
    def check(node, low, high):
        if not node:
            return True
        if not (low < node.val < high):
            return False
        return check(node.left, low, node.val) and check(node.right, node.val, high)

    return check(root, float('-inf'), float('inf'))

root1 = TreeNode(2, TreeNode(1), TreeNode(3))
print(f"Input: [2, 1, 3], Output: {is_valid_bst(root1)}")

root2 = TreeNode(5, TreeNode(6), TreeNode(7))
print(f"Input: [5, 6, 7], Output: {is_valid_bst(root2)}")
```

```python
#Q6 
def binary_search_recursive(arr: list[int], target: int, left=0, right=None) -> int:
    # START OF YOUR CODE
    # Task: Implement a recursive binary search to find the target in the sorted array.
    # Return the index of the target if found, otherwise return -1.
    # Note: Your core search logic MUST be recursive and MUST NOT use loops (for/while).
    if right is None:
        right = len(arr) - 1
    if left > right:
        return -1

    mid = (left + right) // 2

    if arr[mid] == target:
        return mid
    elif arr[mid] < target:
        return binary_search_recursive(arr, target, mid + 1, right)
    else:
        return binary_search_recursive(arr, target, left, mid - 1)
    # END OF YOUR CODE


# Example calls for testing
print(f"Input: arr = [1, 5, 8, 12, 16, 23, 28], target = 12, Output: {binary_search_recursive([1, 5, 8, 12, 16, 23, 28], 12)}")
print(f"Input: arr = [10, 20, 30, 40, 50], target = 25, Output: {binary_search_recursive([10, 20, 30, 40, 50], 25)}")
```

```python
#Q7.Increment large integer digit

def increment_digits(digits: list[int]) -> list[int]:
    result = []
    carry = 1  
    
    
    for i in range(len(digits) - 1, -1, -1):
        total = digits[i] + carry
        if total == 10:
            result.append(0)
            carry = 1
        else:
            result.append(total)
            carry = 0
    
    
    if carry == 1:
        result.append(1)
    
    result.reverse()
    return result
print(f"Input: [1,2,3], Output: {increment_digits([1,2,3])}")
print(f"Input: [9,9], Output: {increment_digits([9,9])}")
```

```python
#Q8 Valid Parenthesis 

def is_valid(s: str) -> bool:
    
    stack = []
    
    bracket_map = {')': '(', ']': '[', '}': '{'}
    
    opening_brackets = set(['(', '[', '{'])
    
    for char in s:
        if char in opening_brackets:
            stack.append(char)
        else:
            if not stack or stack[-1] != bracket_map[char]:
                return False
            stack.pop()
    
    return len(stack) == 0


# Example calls for testing
print(f'Input: "()", Output: {is_valid("()")}')
print(f'Input: "()[]", Output: {is_valid("()[]{}")}')

    
```

```python
#Q10 PRINT FIRST N PRIME NO.

def find_first_n_primes(n: int) -> list[int]:
    # START OF YOUR CODE
    # Task: Generate a list of the first 'n' prime numbers.
    # Store the final answer in the 'prime_numbers' variable.
    prime_numbers = []

    def is_prime(x):
        if x < 2:
            return False
        for i in range(2, int(x**0.5) + 1):
            if x % i == 0:
                return False
        return True

    num = 2
    while len(prime_numbers) < n:
        if is_prime(num):
            prime_numbers.append(num)
        num += 1
    # END OF YOUR CODE
    return prime_numbers

print(f"Input: n = 1, Output: {find_first_n_primes(1)}")
print(f"Input: n = 5, Output: {find_first_n_primes(5)}")
```

```python
#Q12 Product matrices
def multiply_matrices(matrix_a: list[list[float]], matrix_b: list[list[float]]) -> list[list[float]]:
    
    a_rows = len(matrix_a)
    a_cols = len(matrix_a[0]) if a_rows > 0 else 0
    b_rows = len(matrix_b)
    b_cols = len(matrix_b[0]) if b_rows > 0 else 0
    
    
    if a_rows == 0 or a_cols == 0 or b_rows == 0 or b_cols == 0:
        return []
    

    if a_cols != b_rows:
        return []
    
    
    product_matrix = [[0.0 for _ in range(b_cols)] for _ in range(a_rows)]
    
    
    for i in range(a_rows):
        for j in range(b_cols):  
            for k in range(a_cols):  
                product_matrix[i][j] += matrix_a[i][k] * matrix_b[k][j]
    
    return product_matrix
matrix_a_1 = [[1.0, 2.0], [3.0, 4.0]]
matrix_b_1 = [[5.0, 6.0], [7.0, 8.0]]
print(f"Input matrix_a: {matrix_a_1}, matrix_b: {matrix_b_1}, Output: {multiply_matrices(matrix_a_1, matrix_b_1)}")

matrix_a_2 = [[1.0, 2.0, 3.0]]
matrix_b_2 = [[4.0], [5.0]]
print(f"Input matrix_a: {matrix_a_2}, matrix_b: {matrix_b_2}, Output: {multiply_matrices(matrix_a_2, matrix_b_2)}")
```

```python
#Q13 Unique Single No.
def find_single_number(nums: list[int]) -> int:
    # START OF YOUR CODE
    # Task: Find the unique element in 'nums' that appears only once.
    # Store the final answer in the 'single_number' variable.
    single_number = 0
    for n in nums:
        single_number ^=n
    # END OF YOUR CODE
    return single_number

print(f"Input: [2, 2, 1], Output: {find_single_number([2, 2, 1])}")
print(f"Input: [4, 1, 2, 1, 2], Output: {find_single_number([4, 1, 2, 1, 2])}")
```

```python
#Q14 Check Straight Line
def check_straight_line(coordinates: list[list[int]]) -> bool:
    # Agar 2 ya 1 points hai to always straight line
    if len(coordinates) <= 2:
        return True
    
    # First two points leke slope comparison karte hai
    x0, y0 = coordinates[0]
    x1, y1 = coordinates[1]
    
    # Har point ko check karo ki first two ke saath collinear hai ya nahi
    for i in range(2, len(coordinates)):
        x, y = coordinates[i]
        
        # Collinearity check: (y1-y0)*(x-x0) == (y-y0)*(x1-x0)
        if (y1 - y0) * (x - x0) != (y - y0) * (x1 - x0):
            return False
    
    return True
print(f"Input: [[1,1],[2,2],[3,3]], Output: {check_straight_line([[1,1],[2,2],[3,3]])}")
print(f"Input: [[1,1],[2,2],[3,4]], Output: {check_straight_line([[1,1],[2,2],[3,4]])}")
```

```python
#Q15. Ugly Number

def find_nth_ugly_number(n: int) -> int:
    if n <= 0:
        return 0
    
    # First ugly number is 1
    ugly = [0] * n
    ugly[0] = 1
    
    # Pointers for multiples of 2, 3, 5
    i2 = i3 = i5 = 0
    
    for i in range(1, n):
        # Generate candidates
        next2 = ugly[i2] * 2
        next3 = ugly[i3] * 3
        next5 = ugly[i5] * 5
        
        # Find minimum
        next_ugly = min(next2, next3, next5)
        ugly[i] = next_ugly
        
        # Increment pointers which produced the minimum
        # Use if, not elif, because same number can be produced by multiple factors
        if next_ugly == next2:
            i2 += 1
        if next_ugly == next3:
            i3 += 1
        if next_ugly == next5:
            i5 += 1
    
    return ugly[n-1]
print(f"Input: n=1, Output: {find_nth_ugly_number(1)}")
print(f"Input: n=10, Output: {find_nth_ugly_number(10)}") 
```
